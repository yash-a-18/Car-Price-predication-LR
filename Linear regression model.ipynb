# Problem Statement: 
    You will be predicting the costs of used cars given the data collected from various sources and distributed across 
    various locations in India.

### FEATURES:

* `Name`               : The brand and model of the car.

* `Location`           : The location in which the car is being sold or is available for purchase.

* `Year`               : The year or edition of the model.

* `Kilometers_Driven`  : The total kilometres driven in the car by the previous owner(s) in KM.

* `Fuel_Type`: The type of fuel used by the car.

* `Transmission`: The type of transmission used by the car.

* `Owner_Type`: Whether the ownership is Firsthand, Second hand or other.

* `Mileage`: The standard mileage offered by the car company in kmpl or km/kg

* `Engine`: The displacement volume of the engine in cc.

* `Power`: The maximum power of the engine in bhp.

* `Seats`: The number of seats in the car.

* `Price`: The price of the used car in INR Lakhs.

**Splitting Company Name and Model Name**

```python
car_train_dataset['Company_Name'] = car_train_dataset.Name.str.split().str.get(0)
car_test_dataset['Company_Name'] = car_test_dataset.Name.str.split().str.get(0)
```

```python
carModel = (car_train_dataset.Name.str.split(' ').str[1:])
carModel_test = (car_test_dataset.Name.str.split(' ').str[1:])
```

```python
Model = []

for items in carModel:
    tempstr = ""
    for i in range(0,len(items)):
        if(i == 0):
             tempstr = tempstr + items[i]
        else:
             tempstr = tempstr + " " + items[i]
    Model.append(tempstr)
    
car_train_dataset['Model'] = Model
```

```python
Model = []

for items in carModel_test:
    tempstr = ""
    for i in range(0,len(items)):
        if(i == 0):
             tempstr = tempstr + items[i]
        else:
             tempstr = tempstr + " " + items[i]
    Model.append(tempstr)

car_test_dataset['Model'] = Model
```

```python
car_train_dataset = car_train_dataset.drop(['Name'], axis = 1)
car_test_dataset = car_test_dataset.drop(['Name'], axis = 1)
```

```python
car_train_dataset = car_train_dataset[[
    'Company_Name',
    'Model',
    'Location',
    'Year',
    'Kilometers_Driven',
    'Fuel_Type',
    'Transmission',
    'Owner_Type',
    'Mileage',
    'Engine',
    'Power',
    'Seats',
    'Price']]

car_test_dataset = car_test_dataset[[
    'Company_Name',
    'Model',
    'Location',
    'Year',
    'Kilometers_Driven',
    'Fuel_Type',
    'Transmission',
    'Owner_Type',
    'Mileage',
    'Engine',
    'Power',
    'Seats']]
```

```python
car_train_dataset.head()
```

**Separating Value/Number from `Mileage`,`Engine`,`Power`.**

```python
car_train_dataset['Mileage'] = (car_train_dataset['Mileage'].str.split(' ').str[0])
car_train_dataset['Power'] = (car_train_dataset['Power'].str.split(' ').str[0])
car_train_dataset['Engine'] = (car_train_dataset['Engine'].str.split(' ').str[0])

car_test_dataset['Mileage'] = (car_test_dataset['Mileage'].str.split(' ').str[0])
car_test_dataset['Power'] = (car_test_dataset['Power'].str.split(' ').str[0])
car_test_dataset['Engine'] = (car_test_dataset['Engine'].str.split(' ').str[0])
```

I noticed that the dataset have 0.0. So, replacing null values with 0.0.

```python
car_train_dataset["Power"].replace({"null": "0.0",}, inplace=True)
car_train_dataset["Mileage"].replace({"null": "0.0",}, inplace=True)
car_train_dataset["Engine"].replace({"null": "0.0",}, inplace=True)

car_test_dataset["Power"].replace({"null": "0.0",}, inplace=True)
car_test_dataset["Mileage"].replace({"null": "0.0",}, inplace=True)
car_test_dataset["Engine"].replace({"null": "0.0",}, inplace=True)
```

**Converting datatype**

```python
car_train_dataset['Mileage'] = car_train_dataset['Mileage'].astype('float')
car_train_dataset['Power'] = car_train_dataset['Power'].astype('float')
car_train_dataset['Engine'] = car_train_dataset['Engine'].astype('float')

car_test_dataset['Mileage'] = car_test_dataset['Mileage'].astype('float')
car_test_dataset['Power'] = car_test_dataset['Power'].astype('float')
car_test_dataset['Engine'] = car_test_dataset['Engine'].astype('float')
```

## Replacing Null and Unwanted Values

```python
car_train_dataset['Seats'].fillna(2,inplace=True)
car_test_dataset['Seats'].fillna(2,inplace=True)

car_train_dataset['Seats'] = car_train_dataset['Seats'].replace(0,2)
car_test_dataset['Seats'] = car_test_dataset['Seats'].replace(0,2)
```

```python
car_train_dataset['Mileage'].fillna(car_train_dataset['Mileage'].mean(),inplace=True)
car_test_dataset['Mileage'].fillna(car_test_dataset['Mileage'].mean(),inplace=True)

car_train_dataset['Mileage'] = car_train_dataset['Mileage'].replace(0.0, car_train_dataset['Mileage'].mean())
car_test_dataset['Mileage'] = car_test_dataset['Mileage'].replace(0.0, car_train_dataset['Mileage'].mean())
```

```python
car_train_dataset['Engine'].fillna(car_train_dataset['Engine'].mean(),inplace=True)
car_test_dataset['Engine'].fillna(car_test_dataset['Engine'].mean(),inplace=True)

car_train_dataset['Engine'] = car_train_dataset['Engine'].replace(0.0, car_train_dataset['Engine'].mean())
car_test_dataset['Engine'] = car_test_dataset['Engine'].replace(0.0, car_train_dataset['Engine'].mean())
```

```python
car_train_dataset['Power'].fillna(car_train_dataset['Power'].mean(),inplace=True)
car_test_dataset['Power'].fillna(car_test_dataset['Power'].mean(),inplace=True)

car_train_dataset['Power'] = car_train_dataset['Power'].replace(0.0, car_train_dataset['Power'].mean())
car_test_dataset['Power'] = car_test_dataset['Power'].replace(0.0 ,car_train_dataset['Power'].mean())
```

# Linear Regression

```python
from sklearn import preprocessing
```

```python
x = car_train_dataset[['Year','Kilometers_Driven','Mileage','Engine','Power','Seats']]
x.head(5)
```

```python
y = car_train_dataset['Price']
```

```python
#applying train-test-split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=4)
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
lr = LinearRegression().fit(x_train,y_train)
lr
```

```python
print('Intercept: ',lr.intercept_)
print('Coefficients: ',lr.coef_)
```

```python
yhat = lr.predict(x_test)
yhat
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, yhat))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, yhat))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, yhat)))
print('Accuracy:',lr.score(x_test,y_test))
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': yhat})
```

**Actual Vs Predicted**

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
print (f' Train Score is {lr.score(x_train, y_train)}')
print (f' Test Score is {lr.score(x_test, y_test)}')
```

```python
#applying gradient boosting
from sklearn.ensemble import GradientBoostingRegressor
xTrain, xTest, yTrain, yTest = train_test_split(x, y,train_size = 0.80)

gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=10, random_state=0, loss='huber')
gb = gb.fit(xTrain, yTrain)
y_pred = gb.predict(xTest)
y_pred
print("Accuracy:{0:.3f}".format(gb.score(xTest,yTest)))
```

```python
print (f' Train Score is {lr.score(xTrain, yTrain)}')
print (f' Test Score is {lr.score(xTest, yTest)}')
```
